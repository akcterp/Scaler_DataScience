```python
df1 = pd.DataFrame({'name':["Ram","Shyam","Mukesh","Suresh"],'age':[10,18,60,50]})
```

```python
df1
```

```python
1. If age<18: kid
2. age >=18 and <=50: adult
3. age>50 senior
```

```python
df2 = df1.copy()
df1.loc[df2["age"]<18,"age"] = "kid"
df1.loc[df2["age"]>50] = "senior"
df1.loc[(df2["age"]>=18) & (df2["age"]<=50)] = "adult"
```
